```python
# importing modules
import io
import pandas as pandas
from matplotlib import pyplot as plt
```

```python
# using the class method to construct the data visualization model
# this module will be analysing and working on country_wise_latest csv file.
class VisualiseModel:
    def openFile(self):
        self.datafile = pandas.read_csv("country_wise_latest.csv")
        # print(self.datafile)
    
    # Data fetching for India
    def generateIndiaData(self):
        # fetching out some labels for our analysis
        self.indiaData = self.datafile[self.datafile.country == "India"]
        # print(self.indiaData)
        self.plotIndiaGraph()
    
    def plotIndiaGraph(self):
        # plotting the graph index for India
        plt.hist(self.indiaData.confirmed, self.indiaData.recovered)
        plt.hist(self.indiaData.confirmed, self.indiaData.deaths)
        plt.title("Covid 19 Indian Index")
        plt.xlabel("Confirmed Cases")
        plt.ylabel("Cured cases & Death cases")
        plt.legend(["Cured", "Deaths"])
        plt.show()

    # Data fetching for Australia
    def generateAustraliaData(self):
        self.australiaData = self.datafile[self.datafile.country == "Australia"]
        # print(self.australiaData)
        self.plotAustraliaData()
    
    def plotAustraliaData(self):
        # plotting the graph index for Australia
        plt.hist(self.australiaData.confirmed, self.australiaData.recovered)
        plt.hist(self.australiaData.confirmed, self.australiaData.deaths)
        plt.title("Covid 19 Australian Index")
        plt.xlabel("Confirmed Cases")
        plt.ylabel("Cured cases & Death cases")
        plt.legend(["Cured"], "Deaths")
        plt.show()
```

```python
# teser method for out code module
def tester():
    visualiseModel = VisualiseModel()
    visualiseModel.openFile()
    # testing India data
    visualiseModel.generateIndiaData()
    # testing australia data
    visualiseModel.generateAustraliaData()
```

```python
# this is the main function which will be calling tester function to test out the program
if __name__ == '__main__':
    tester()
```
